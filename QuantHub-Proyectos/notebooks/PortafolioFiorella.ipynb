```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import numpy_financial as npf
```

```python
# Datos
escenarios = ["Depsito+ETF", "Alquiler Optimista", "Alquiler Realista", "Alquiler Pesimista"]
ganancias = [3990, 25800, 21400, 16000]  # Ganancia neta en 3 aos

# Grfico
plt.figure(figsize=(8,6))
bars = plt.bar(escenarios, ganancias)

# Colores para diferenciar
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']
for bar, color in zip(bars, colors):
    bar.set_color(color)

# Etiquetas de valores encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 300, f"S/ {yval:,}", 
             ha='center', va='bottom', fontsize=10)

# Esttica
plt.title("Comparacin de Ganancia Neta en 3 aos", fontsize=14, weight='bold')
plt.ylabel("Ganancia neta (S/)", fontsize=12)
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
